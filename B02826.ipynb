```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('bezdekIris.csv')

# Rename columns
new_column_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
df.columns = new_column_names

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Group the data by 'Class' (species)
grouped = df.groupby('Class')

# Compute mean values for each feature grouped by species
mean_values = grouped.mean()
print("\nMean values for each feature grouped by species:")
print(mean_values)

# Compute median values for each feature grouped by species
median_values = grouped.median()
print("\nMedian values for each feature grouped by species:")
print(median_values)

# Compute variance values for each feature grouped by species
variance_values = grouped.var()
print("\nVariance values for each feature grouped by species:")
print(variance_values)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('bezdekIris.csv')

# Rename columns
new_column_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
df.columns = new_column_names

# List of unique species
species = df['Class'].unique()

# Set the style of seaborn
sns.set(style="white")

# Create a plot for each species
plt.figure(figsize=(18, 12))

for i, specie in enumerate(species):
    plt.subplot(2, 2, i + 1)
    species_df = df[df['Class'] == specie]
    correlation_matrix = species_df[['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1, center=0, square=True)
    plt.title(f'Correlation Matrix for {specie}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('bezdekIris.csv')

# Rename columns
new_column_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
df.columns = new_column_names
df['sepal_ratio'] = df['Sepal_Length'] / df['Sepal_Width']
df['petal_ratio'] = df['Petal_Length'] / df['Petal_Width']
grouped = df.groupby('Class').agg({
    'sepal_ratio': 'mean',
    'petal_ratio': 'mean'
})
max_sepal_ratio = grouped['sepal_ratio'].idxmax()
max_petal_ratio = grouped['petal_ratio'].idxmax()
print("Average sepal_ratio by species:")
print(grouped['sepal_ratio'])
print("\nSpecies with the largest average sepal_ratio:", max_sepal_ratio)

print("\nAverage petal_ratio by species:")
print(grouped['petal_ratio'])
print("\nSpecies with the largest average petal_ratio:", max_petal_ratio)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('bezdekIris.csv')

# Rename columns
new_column_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
df.columns = new_column_names

# Create new features: sepal_ratio and petal_ratio
df['sepal_ratio'] = df['Sepal_Length'] / df['Sepal_Width']
df['petal_ratio'] = df['Petal_Length'] / df['Petal_Width']
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 1, figsize=(12, 10), sharex=True)
sns.histplot(data=df, x='sepal_ratio', hue='Class', multiple='stack', bins=20, ax=axes[0])
axes[0].set_title('Distribution of Sepal Ratio by Species')
axes[0].set_xlabel('Sepal Ratio')
axes[0].set_ylabel('Frequency')
sns.histplot(data=df, x='petal_ratio', hue='Class', multiple='stack', bins=20, ax=axes[1])
axes[1].set_title('Distribution of Petal Ratio by Species')
axes[1].set_xlabel('Petal Ratio')
axes[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('bezdekIris.csv')

# Rename columns
new_column_names = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
df.columns = new_column_names

# Create new features: sepal_ratio and petal_ratio
df['sepal_ratio'] = df['Sepal_Length'] / df['Sepal_Width']
df['petal_ratio'] = df['Petal_Length'] / df['Petal_Width']

# Features to use for PCA
features = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'sepal_ratio', 'petal_ratio']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the PCA results and species
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Class'] = df['Class']

# Plot the PCA results
plt.figure(figsize=(10, 8))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Class', palette='viridis', s=100)
plt.title('PCA of Iris Dataset (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```
